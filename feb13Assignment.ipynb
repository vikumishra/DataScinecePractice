### 3)

Errors defined in ArithmeticError class are - FloatingPointError , ZeroDivisionError,OverflowError etc.



```python
# ZeroDivisionError=> It occurs when we try to divide by zero

a=10
try:
    a/0
except ZeroDivisionError as e:
    print(e)
```

```python
# OverflowError - Itoccurs when any operations like arithmetic operations or any 
# other variable storing any value above its limit
import math
try:
    print(math.exp(10000))
except OverflowError as e:
    print(e)
    
```
